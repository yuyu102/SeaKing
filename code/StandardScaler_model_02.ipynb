```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/2011_2020_m2.ver2.csv', encoding='utf-8')
df
```

```python
print(df["특보강도"].value_counts())
```

```python
df_tgt = df[['WIND_SPEED', 'GUST_WIND', 'SIG_WAVE', 'MAX_WAVE', 'HPA']]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df_tgt)
```

```python
arr = scaler.transform(df_tgt)
```

```python
df1 = pd.DataFrame(arr, columns=df_tgt.columns)
df1
```

```python
df1 = pd.concat([df1,df[['일시','특보강도','남은시간']]], axis=1)
df1
```

```python
print('feature 들의 평균 값')
print(df1.mean())
print('\nfeature 들의 분산 값')
print(df1.var())
```

```python
df1['일시'] = pd.to_datetime(df1['일시'])
df1.info()
```

```python
df1 = df1.set_index('일시')
df1
```
