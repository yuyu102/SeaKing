```python
# 라이브러리 로드
import pandas as pd
import numpy as np
```

```python
# 데이터 로딩
df = pd.read_csv('./data/2011_2020_m2.ver2.csv', encoding='utf-8')
df
```

```python
# 데이터셋 구성
df_tgt = df[['WIND_SPEED', 'GUST_WIND', 'SIG_WAVE', 'MAX_WAVE', 'HPA', '특보강도']]
```

```python
print('feature들의 최소 값')
print(df_tgt.min())
print('\nfeature들의 최대 값')
print(df_tgt.max())
```

* MinMax

```python
# MinMax 라이브러리 로드
from sklearn.preprocessing import MinMaxScaler
```

```python
# MinMaxScaler객체 생성
scaler = MinMaxScaler()
```

```python
scaler.fit(df_tgt)
```

```python
arr = scaler.transform(df_tgt)
```

```python
arr
```

```python
df1 = pd.DataFrame(arr, columns=df_tgt.columns)
df1
```

```python
# concat
df1 = pd.concat([df1,df[['일시','남은시간']]], axis=1)
df1
```

```python
print('feature 들의 평균 값')
print(df1.mean())
print('\nfeature 들의 분산 값')
print(df1.var())
```

```python
# obj -> datatime
df1['일시'] = pd.to_datetime(df1['일시'])
df1.info()
```

```python
# '일시' index
df1 = df1.set_index('일시')
df1
```
